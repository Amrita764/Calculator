```python
# Step 2: Generate synthetic dataset
X, _ = make_blobs(n_samples=300, centers=1, cluster_std=0.5, random_state=42)

# Add some anomalies (outliers)
np.random.seed(42)
outliers = np.random.uniform(low=-6, high=6, size=(20, 2))
X = np.vstack((X, outliers))

# Visualize dataset
plt.figure(figsize=(6,5))
plt.scatter(X[:, 0], X[:, 1], c='blue', label='Data Points')
plt.title("Dataset with Some Anomalies")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
# Step 3: Create Isolation Forest model
iso_forest = IsolationForest(contamination=0.05, random_state=42)

# Fit the model
iso_forest.fit(X)

# Predict anomalies (-1 = anomaly, 1 = normal)
y_pred = iso_forest.predict(X)
```

```python
# Step 1: Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.datasets import make_blobs
```

```python
# Step 4: Visualize normal points vs anomalies
plt.figure(figsize=(6,5))
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], c='blue', label='Normal')
plt.scatter(X[y_pred == -1, 0], X[y_pred == -1, 1], c='red', label='Anomaly')
plt.title("Anomaly Detection using Isolation Forest")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
# Step 5: Count results
normal = list(y_pred).count(1)
anomaly = list(y_pred).count(-1)

print(f" Normal points: {normal}")
print(f"  Anomalies detected: {anomaly}")
```

```python
# Load dataset
data = pd.read_csv("https://raw.githubusercontent.com/nsethi31/Full-Stack-Data-Science-with-Python/master/Data/creditcard.csv")

# Use numerical columns only
X = data[['Time', 'V1', 'V2', 'V3', 'Amount']]

# Build model
iso_forest = IsolationForest(contamination=0.02, random_state=42)
y_pred = iso_forest.fit_predict(X)

# Add prediction column
data['Anomaly'] = y_pred
print(data['Anomaly'].value_counts())
```

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import IsolationForest
import pandas as pd

# Load the Iris dataset
iris = load_iris()
X_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

# Build and fit the Isolation Forest model
iso_forest_iris = IsolationForest(contamination=0.05, random_state=42)
y_pred_iris = iso_forest_iris.fit_predict(X_iris)

# Add prediction column to the Iris DataFrame
X_iris['Anomaly'] = y_pred_iris

# Display the value counts of the anomaly column
print(X_iris['Anomaly'].value_counts())
```
